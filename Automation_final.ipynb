```python
listOflinks=[]   #Create a list in which all the links will be appended
condition = 1    #Condition for getting certain number of pages
condition1=True #Youtube baba says do this and it actually works :')
a=int(input())  #User defined number of pages
while condition<=a:   
    UniIntoList= webD.find_elements_by_class_name('school-card')    
    for el in UniIntoList:
        listOflinks.append(el.get_property('href'))             #for all the class names get the href and append it into listOflinks
    try:
        webD.find_elements_by_class_name('page-link')[-1].click()  #Pagination
        a=a-1
    except: 
        condition1=False
  
```

```python
listOflinks1 = list(dict.fromkeys(listOflinks))   # Just to remove duplicates in the list
listOflinks1
```

```python
from tqdm import tqdm                       #To check the process
alldetails=[]                               #new list
for i in tqdm(listOflinks1):
    webD.get(i)
    nameoftheschool=webD.find_element_by_xpath('//*[@id="college_name"]').text
    schooldescription=webD.find_element_by_xpath('//*[@id="school_description"]').text
    priceoftheschool=webD.find_element_by_xpath('//*[@id="avg_tuition_outstate"]').text
    acceptancerate=webD.find_element_by_xpath('//*[@id="acceptance_rate"]').text
    avgsalary=webD.find_element_by_xpath('//*[@id="salary_ten"]').text
    ischool_rank=webD.find_element_by_xpath('//*[@id="ischool_rank"]').text
    tempJ={'nameoftheschool':nameoftheschool,
      'priceoftheschool':priceoftheschool,
      'schooldescription':schooldescription,
      'acceptancerate':acceptancerate,
      'avgsalary':avgsalary,
      'ischool_rank':ischool_rank,    
      'linkofProduct':i}
    alldetails.append(tempJ)
```
